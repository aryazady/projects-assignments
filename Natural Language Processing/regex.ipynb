```python
import json
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import re
from hazm import *
import pandas as pd
```

**Crawler to Get Text**

```python
def get_page_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
    else:
        print(f"Error accessing {url}")
        return

    soup = BeautifulSoup(html, 'html.parser')

    # Get title
    title_div = soup.find('div', id='page-hierarchy')
    title = title_div.h2.a.get_text()
    title = title.replace('\r', '')
    title = re.sub('\n', ' ', title)
    title = re.sub('\s+', ' ', title).strip()
    try:
        title = title.split("-")[1].strip()
    except:
        title = title

    # Get text
    divs = soup.find_all('div', class_='n')
    text = []
    for div in divs:
        paragraphs = div.find_all('p')
        for para in paragraphs:
            text.append(para.get_text()+ "\n")
    text = ' '.join(text)
    return {'title': title, 'text': text}

result1 = []

print("Crawling first baab...")
with tqdm(total=41) as pbar:
    for i in range(1, 41):
        url = f"https://ganjoor.net/saadi/golestan/gbab1/sh{i}"
        data = get_page_data(url)
        if data:
            result1.append(data)
        pbar.update(1)

with open('golestan_bab1.json', 'w') as f:
    json.dump(result1, f, ensure_ascii=False, indent=4)

result2 = []

print("Crawling second baab...")
with tqdm(total=48) as pbar:
    for i in range(1, 48):
        url = f"https://ganjoor.net/saadi/golestan/gbab2/sh{i}"
        data = get_page_data(url)
        if data:
            result2.append(data)
        pbar.update(1)

with open('golestan_bab2.json', 'w') as f:
    json.dump(result2, f, ensure_ascii=False, indent=4)

result3 = []

print("Crawling third baab...")
with tqdm(total=28) as pbar:
    for i in range(1, 28):
        url = f"https://ganjoor.net/saadi/golestan/gbab3/sh{i}"
        data = get_page_data(url)
        if data:
            result3.append(data)
        pbar.update(1)


with open('golestan_bab3.json', 'w') as f:
    json.dump(result3, f, ensure_ascii=False, indent=4)


```
